## Consider calculating the fibonacci sequence:

### Pure recursion:


```
int fib(int x)
{
    if (x < 2)
        return 1;

    return fib(x-1) + fib(x-2);
}
```

### Recursion with memoization/DP:



```
void fib(int x)
{
    static vector<int> cache(N, -1);

    int& result = cache[x];

    if (result == -1)
    {
        if (x < 2)
            result = 1;
        else
            result = fib(x-1) + fib(x-2);
    }

    return result;
}
```

### The following would also be considered DP, but without recursion:

```
int fibresult[N];

void setup_fib()
{
    fibresult[0] = 1;
    fibresult[1] = 1;
    for (int i = 2; i < N; i++)
       fibresult[i] = fibresult[i-1] + fibresult[i-2];
}

int fib(int x) { return fibresult[x]; }
```

###Also the following is neither recursion nor DP:

```
int fib(int x)
{
    int a = 1;
    int b = 1;
    for (int i = 2; i < x; i++)
    {
        a = a + b;
        swap(a,b);
    }
    return b;
}
```





````
# Pure recursion:
%%time
#n=int(input('enter the nummber n = '))
n=5
def fact(n):
  if(n<1):
    return 1
  return n*fact(n-1)
  
print('Factorial of given no. is : ',fact(n))
````

````
#Recursion with memoization/DP:
%%time
#n=int(input('enter the nummber n = '))
fl=[]
ans=0
n=5
def fact(n):
  if(n<1):
    return 1
  else:
    ans = n*fact(n-1)
    fl.append(ans)
    return ans
print('Factorial of given no. is : ',fact(n))
````

````
# DP, but without recursion:

%%time
ans =[]
#ans.append(5)
#n=int(input('enter the nummber n = '))

fact = 1
n=5

for i in range(1,n+1):
  fact*=i
  if fact in ans:
    fact= ans[ans.index(ans)]*i
    ans.append(fact)
  else:
    ans.append(fact)
print('Factorial of given no. is : ',ans[-1])
````
